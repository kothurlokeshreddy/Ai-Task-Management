## Preprocessing to task_description column

```python
def preprocess_text(text):
    # 1. Normalize: Lowercase & remove punctuation
    text = text.lower()
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    
    # 2. Tokenize
    tokens = nltk.word_tokenize(text)
    
    # 3. Remove stopwords
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # 4. Apply Stemming
    stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    
    # Return Joined and cleaned Text
    return " ".join(stemmed_tokens)

df['cleaned_description'] = df['task_description'].astype(str).apply(preprocess_text)
```

# Feature Extraction using TF-IDF

##### TF-IDF : TF-IDF (Term Frequency-Inverse Document Frequency) is a numerical statistic used in NLP to reflect the importance of a word to a document in a collection or corpus. It's a technique that converts text into numerical vectors by multiplying term frequency (TF) with inverse document frequency (IDF).

##### How TF-IDF Works:
       Term Frequency (TF): Measures how often a term appears in a document. 
       Formula: TF(t, d) = (Number of times term t appears in document d) / (Total number of terms in document d) 
       Inverse Document Frequency (IDF): Measures how important a term is across the entire corpus. 
       Formula: IDF(t) = log\_e (Total number of documents / Number of documents with term t in it) 
##### TF-IDF: The product of TF and IDF, indicating the term's importance within a specific document relative to the entire corpus. 
##### Formula: TF-IDF(t, d) = TF(t, d) * IDF(t)

### Importing Libraries

```python
!pip install -q sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
import nltk
import pandas as pd
```

## 1. TF-IDF Implementation

```python
# Using Cleaned Text
descriptions = df['task_description'].fillna("")

# Initialize and TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features = 5000)
x_tfidf = tfidf.fit_transform(descriptions)

#convert to dataframe
tfidf_df = pd.DataFrame(x_tfidf.toarray(), columns = tfidf.get_feature_names_out())


tfidf_df.head()
```

#### TF (Term Frequency): Counts how often a word appears.

#### IDF (Inverse Document Frequency): Penalizes common words.

#### TF-IDF highlights important but unique words.

## 2. Word2Vec Embedding

#### Each word is mapped to a vector using word context (semantic meaning).

```python
# Tokenize descriptions for Word2Vec
tokenized_text = descriptions.apply(nltk.word_tokenize)


# Train Word2Vec model
w2v_model = Word2Vec(sentences = tokenized_text, vector_size = 100, window = 5, min_count = 2, workers = 4)


# Function to convert description to vector (mean of word vectors)
def get_vector(doc):
    if not doc:
        return np.zeroes(100)
    
    vectors = [w2v_model.wv[word] for word in doc if word in w2v_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)


# Apply on entire dataset
df['w2v_vector'] = tokenized_text.apply(get_vector)


# Example: view first vector
print("Vector shape:", df['w2v_vector'][0].shape)
print(df['w2v_vector'][0])
```

## 3. BERT Embedding (Transformer-Based Model)
#### BERT gives context-aware embeddings using pre-trained deep models.

```python
from sentence_transformers import SentenceTransformer

# Load pre-trained BERT model
bert_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate BERT embeddings for each task description
bert_embeddings = bert_model.encode(descriptions.tolist(), show_progress_bar=True)

# Convert to DataFrame
bert_df = pd.DataFrame(bert_embeddings)
print("BERT Embedding shape:", bert_df.shape)
print(bert_df.head(10))
```

## Implementation of Random Forest and XG Boost

```python
# Encoding Task_type column and assigned_to column
from sklearn.preprocessing import LabelEncoder, StandardScaler

task_type_le = LabelEncoder()
df["task_type_encoded"] = task_type_le.fit_transform(df['task_type'].astype(str))
df["assigned_to_encoded"] = task_type_le.fit_transform(df['assigned_to'].astype(str))


# Combining Numeric + Encoded Features
X_numeric = df[["estimated_hours", "actual_hours", "days_to_deadline", "is_overdue",
                "task_type_encoded", "status_encoded", "assigned_to_encoded"]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

# Combine TF-IDF text features with scaled numeric features
X_combined = hstack([x_tfidf, X_scaled])

# Define target and split data
y_rf = df["priority_encoded"]
X_rf_train, X_rf_test, y_rf_train, y_rf_test = train_test_split(X_combined, y_rf, test_size=0.2, random_state=42)


# === RANDOM FOREST ===
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_rf_train, y_rf_train)
rf_preds = rf_model.predict(X_rf_test)

# === XGBOOST ===
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb_model.fit(X_rf_train, y_rf_train)
xgb_preds = xgb_model.predict(X_rf_test)


# === Evaluation Function ===
def evaluate_model(name, y_true, y_pred):
    print(f"\n=== {name} Evaluation ===")
    print(f"Accuracy: {accuracy_score(y_true, y_pred) * 100:.2f}%")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Evaluate models
evaluate_model("Random Forest", y_rf_test, rf_preds)
evaluate_model("XGBoost", y_rf_test, xgb_preds)
```

# Workload Balancing using Heuristic Approach

#### We use estimated_hours, actual_hours, deadline, and priority as balancing factors
#### Assign Critical/High priority tasks to users with the least total workload
#### User-wise total current workload (sum of estimated_hours of their tasks)

```python
# Map priority to numerical score
priority_weight = {
    "Critical": 3,
    "High": 2,
    "Medium": 1,
    "Low": 0.5
}
df["priority_score"] = df["priority"].map(priority_weight)

# Sort tasks by priority and urgency
task_pool = df.copy().sort_values(by=["priority_score", "days_to_deadline"], ascending=[False, True])

# Initialize workload per user
users = df["assigned_to"].dropna().unique()
user_workload = {user: 0 for user in users}

# Assign each task to least-loaded user
balanced_assignments = []
for _, task in task_pool.iterrows():
    user = min(user_workload, key=user_workload.get)
    balanced_assignments.append(user)
    user_workload[user] += task["estimated_hours"]

# Add balanced assignments to dataset
task_pool["balanced_assigned_to"] = balanced_assignments

# Group by new assignment and summarize workload
workload_summary = task_pool.groupby("balanced_assigned_to")["estimated_hours"].sum().sort_values(ascending=False)

# Return preview and workload balance summary
task_pool[["task_id", "priority", "estimated_hours", "balanced_assigned_to"]].head(), workload_summary.head()
```
